See how to specify the color palette(note in scatter_mapbox the color need to specify in hex, so we need to parse the color to hex by some sort)
TIPS: While Plotly has its own palette, but for python, I found the easy way to select a palette is using seaborn sns.color_palette(), all the available pallet can be found in colorbrewer.

```python
from matplotlib.colors import to_hex
import seaborn as sns
# px.set_mapbox_access_token(mapbox_access_token)
fig = px.scatter_mapbox(df_plot_tmp, lat="LATITUDE", lon="LONGITUDE", color="COAST_50KM", 
                         size="QUAL_POP",size_max=10,
                        category_orders={'age_cate':['A22-', 'A23-26', 'A27-28', 'A29-31', 'A32-40', 'A40+']},
                        color_discrete_sequence=[to_hex(c) for c in sns.color_palette('BrBG_r', 6)],  
                        
                        zoom=3.5,
                        center=dict(lon=25.931227857669064, lat=50.1021642983885728),
                        width=900, height=600,)
fig.show()
```

Next, play with the continuous/numerical color.

```python
# play with the color: continous/numerical
# px.set_mapbox_access_token(mapbox_access_token)
fig = px.scatter_mapbox(df_plot_tmp, lat="LATITUDE", lon="LONGITUDE", color="COAST_50KM", 
                         size="QUAL_POP",
                        color_continuous_scale=px.colors.cyclical.Twilight, 
                        zoom=4.5,
                        center=dict(lon=25.931227857669064, lat=50.1021642983885728),
                       width=900,
                        height=600,)
fig.show()
```
